```python
conditionResultList = []
def SignRetroReflectivity(df, conditionResultList):
  for row in df.iterrows():
    print(row[1][0])
    median = row[1][14]
    std = row[1][15]
    typeOfSign = row[1][2]
    if (typeOfSign == 'Warning Sign'):
      print("Warning Sign")
      #pass conditions 
      if (median >= 0.8 and std <= 0.065):
        conditionResultList.append('PASS')   
      #marginal conditions
      elif (((0.69 <= median) and (median <= 0.72)) and std >= 0.065):
        conditionResultList.append('MARGINAL')
      #fail conditions
      elif (median <= 0.69 and std >= 0.065):
        conditionResultList.append('FAIL')
      else:
        conditionResultList.append("IDK")
    elif (typeOfSign == 'Regulatory Sign'):
      print("Regulatory Sign")
      #pass conditions 
      if (median >= 0.83 and std <= 0.0498):
        conditionResultList.append('PASS')
      #marginal conditions
      elif (((0.67 <= median) and (median <= 0.68)) and std >= 0.0498):
        conditionResultList.append('MARGINAL')
      #fail conditions
      elif (median <= 0.67 and std >= 0.0498):
        conditionResultList.append('FAIL')
      else:
        conditionResultList.append("IDK")
    elif (typeOfSign == 'Guide Sign'):
      print("Guide Sign")
      #pass conditions 
      if (median >= 0.8 and std <= 0.059):
        conditionResultList.append('PASS')
      #marginal conditions
      elif (((0.69 <= median) and (median <= 0.7)) and std >= 0.059):
        conditionResultList.append('MARGINAL')
      #fail conditions
      elif (median <= 0.69 and std >= 0.059):
        conditionResultList.append('FAIL')
      else:
        conditionResultList.append("IDK")
    else:
      conditionResultList.append("1")
  return conditionResultList
  
```

```python
returnList = SignRetroReflectivity(df, conditionResultList)
```
