```python
allSignsDf = pd.read_csv("/content/finalalgoresult(May1).csv")
allSignsDf
```

```python
allSignIds = allSignsDf.Sign_ID.unique()
```

```python
from numpy import singlecomplex
pathToAllSigns = "/content/allSigns"
for signId in allSignIds:
  signDf = allSignsDf[allSignsDf['Sign_ID'] == signId]
  signDfPath = pathToAllSigns + "/sign" + str(signId) + ".csv"
  signDf.to_csv(signDfPath)
  
```
